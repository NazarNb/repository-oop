### Модифікую програму :додаю своє ім'я до списку та виконую всі завдання з 5 номеру

````python
import random

class MyName:
    """Опис класу / Документація    """
    total_names = 0 # Class Variable
    _default_domain = "itcollege.lviv.ua" # Додана змінна для домену

    def __init__(self, name=None, email_domain=None) -> None:
        """Ініціалізація класу
        - Модифіковано: self.name завжди починається з великої літери.
        - Додано перевірку на цифри/символи.
        - Додана можливість модифікації домену.
        """
        
        # 6. Зміна email_domain: дозволяє модифікувати значення після @
        self._current_domain = email_domain if email_domain is not None else self._default_domain
        
        # --- 7. Перевірка на цифри або символи ---
        if name is not None and str(name).strip() and not str(name).isalpha():
            # Дозволяємо лише літери, ігноруючи None та пусті рядки
            # Якщо name == "Anonymous", також дозволяємо, бо це внутрішній виклик
            if name != "Anonymous":
                raise ValueError("Ім'я може містити лише літери!")

        # 5. Модифікація конструктора: self.name завжди з великої літери
        if name is not None:
            name = str(name).capitalize()

        # Виклик anonymous_user() делегується, якщо name=None
        self.name = name if name is not None else self.anonymous_user().name
        
        MyName.total_names += 1 # modify class variable
        self.my_id = self.total_names

    @property
    def whoami(self) -> str: 
        """Class property        return: повертаємо імя         """
        return f"My name is {self.name}"
    
    @property
    def my_email(self) -> str:
        """Class property        return: повертаємо емейл        """
        return self.create_email()

    # Додана функція в класі, яка рахує кількість букв імені
    @property
    def name_length(self) -> int:
        """Повертає кількість букв імені (використовує len())."""
        return len(self.name)
    
    # Додана нова властивість full_name
    @property
    def full_name(self) -> str:
        """Повертає результат у форматі: "User #<id>: <name> (<email>)" """
        return f"User #{self.my_id}: {self.name} ({self.my_email})"
    
    # 6. Зміна методу create_email для використання модифікованого домену
    def create_email(self) -> str:
        """Instance method        """
        return f"{self.name}@{self._current_domain}"

    # 10. Реалізація методу save_to_file
    def save_to_file(self, filename="users.txt"):
        """ Зберігає рядок із записом у файл. """
        record = self.full_name
        try:
            with open(filename, 'a', encoding='utf-8') as f:
                f.write(record + '\n')
            print(f"[{record}] збережено до файлу {filename}")
        except IOError as e:
            print(f"Помилка при записі у файл: {e}")


    @classmethod
    def anonymous_user(cls):
        """Classs method        """
        return cls("Anonymous")
    
    # 2. Модифікація тексту привітання при виклику методу say_hello()
    @staticmethod
    def say_hello(message="Вітаємо всіх! Hello to everyone!") -> str:
        """Static method        """
        return f"You say: {message}"

# --- ВИКОНАННЯ КОДУ ---
print("Розпочинаємо створювати обєкти!")

# Вхідний список імен (Назар доданий)
names = ("Bohdan", "Marta", "Назар",  None) 
# Створення об'єктів
all_names = {name: MyName(name) for name in names if name is not None}
# Додаткове створення об'єктів для тестування змін
all_names["Анонім"] = MyName(None)
all_names["СпецДомен"] = MyName("Olena", email_domain="special.domain.ua") 

for name, me in all_names.items():
    print(f"""{">*<"*20}
This is object: {me} 
This is object attribute: {me.name} / ID: {me.my_id}
This is {type(MyName.whoami)}: {me.whoami} / {me.my_email}
This is {type(me.create_email)} call: {me.create_email()}
This is static {type(MyName.say_hello)} with defaults: {me.say_hello()} 
***
Нове привітання (модифікований виклик): {me.say_hello('Це нове індивідуальне повідомлення!')}
Довжина імені: {me.name_length}
Повний опис (full_name): {me.full_name}
***
This is class variable {type(MyName.total_names)}: from class {MyName.total_names} / from object {me.total_names}
{"<*>"*20}""")
    me.save_to_file()
print(f"\nWe are done. Total objects created: {MyName.total_names}")
print(f"Кількість імен у списку names: {len(names)}.")
````

1. **Чому при передачі `None` створюється об'єкт з іменем `Anonymous`?**

   Коли в конструктор `__init__` передається `None`, реалізовано делегування: виконується `self.anonymous_user().name`. Метод класу `anonymous_user` створює тимчасовий об'єкт зі значенням "Anonymous" і повертає його ім'я. Через це основному об'єкту присвоюється рядок "Anonymous".
