```python
import random

class Item:
    """Це наш клас який описує ітем в грі"""
    def __init__(self):
        # випадкова назва, але з гарантією, що не порожня
        names = ["Trash", "Axe", "Sword", "Spear", "Bow", "Knife"]
        self.name = random.choice(names) or "Unknown Item"

        # рідкість
        self.rarity = random.choice(["common", "uncommon", "rare", "epic"])

        # гарантуємо непорожні й ненульові значення
        self.attack_power = random.randint(4, 6)
        if self.attack_power <= 0:
            self.attack_power = 1

        self.durability = random.randint(30, 45)
        if self.durability <= 0:
            self.durability = 1

        # випадкова магія з підстраховкою
        magics = ["Магія Природи", "Магія Вітру", "Магія Вогню", "Магія Землі"]
        self.magic_attribute = random.choice(magics) or "Без магії"

        self.description = "Випадковий предмет озброєння"

        # бонус за рідкість
        if self.rarity == "uncommon":
            self.attack_power += 1
        elif self.rarity == "rare":
            self.attack_power += 2
        elif self.rarity == "epic":
            self.attack_power += 4

    def attack(self):
        bonus = random.randint(1, 3)
        attack = self.attack_power + bonus
        return f"{self.name} наніс {attack} шкоди типом {self.magic_attribute}"
    
    def defence(self):
        self.durability += 1
        return f"{self.name} захищається і підсилює міцність до {self.durability}"
        

for i in range(10):
    item = Item()
    print(f"""
Подія {i}: Вам випав предмет:
  Назва: {item.name}
  Рідкість: {item.rarity}
  Параметри:
    Атака/міцність: {item.attack_power}/{item.durability}
    Магічна властивість: {item.magic_attribute}
Опис: {item.description}
""")
    
    if random.randint(0, 1):
        print(f"Застосування: {item.attack()}")
    else:
        print(f"Застосування: {item.defence()}")
```

```python
import random

class Item:
    """Це наш клас який описує предмет у грі"""
    def __init__(self):
        self._initial_durability = random.randint(35, 50) 
        self.name = random.choice(["trash", "axe", "sword", "spear", "bow", "knife"]).capitalize()

        self.rarity = random.choice(["common", "uncommon", "green", "rare"])

        self.attack_power = random.randint(4, 7)

        self.durability = self._initial_durability

        self.magic_attribute = random.choice([
            "Магія Природи",
            "Магія Вогню",
            "Магія Вітру",
            "Магія Каменю"
        ])

        self.description = "Випадково згенерований предмет озброєння"

    @property
    def get_attack_power(self):
        return f"Атака предмета {self.name}: {self.attack_power}"

    @property
    def get_current_durability(self):
        return f"{self.name} має міцність {self.durability}/{self._initial_durability}"
```

### Методи та статичні методи, змінна self

```python
import random

class Item:
    global_item_count = 0  

    """Це наш клас який описує предмет у грі"""
    def __init__(self):
        self._initial_durability = random.randint(35, 55)
        self.name = random.choice(["Hammer", "Axe", "Sword", "Bow", "Knife", "Staff"])

    def test_method(self):
        Item.global_item_count += 1
        return f"{self.name} створено, його міцність: {self._initial_durability}"

    @staticmethod
    def always_say_hello():
        return "Привіт! Я статичний метод класу Item."

# посилання на клас
i = Item
# об'єкт класу
b = Item()

print(type(i), type(b))  # типи різні

print(b.test_method(), i.test_method(b))

print(b.always_say_hello(), i.always_say_hello())
```

### Альтернативні конструктори та класові методи

```python
class Item:
    def __init__(self):
        # тепер для кожного обєкта є характерним дані атрибути
        self._initial_durability = random.randint(30, 45)
        self.name = random.choice(["trash", "Axe", "Sword", "Spear", "Bow", "Knife"])

    @classmethod
    def modify_class_durability(cls, new_durability):
        _a = cls()
        _a._initial_durability = new_durability
        _a.type = "modified"
        print(f"Створено обєкт класу {_a.name} з міцністю {_a._initial_durability}")
        return _a
    a = Item()
b = Item.modify_class_durability(100)
print(a._initial_durability, b.type)
   
```

```python
import random

class Item:
    def __init__(self):
        self.value = random.randint(10, 50) # Зміна імені та діапазону
        self.status = "New" # Новий публічний атрибут
        self._internal_id = f"ID-{self.value}" # Зміна імені захищеного атрибута
        self.__secret_key = f"KEY-{self.value % 5}" # Зміна імені захищеного атрибута
    
    # 2. Змінено функціонал __add__ (тепер множить)
    def __add__(self, other: 'Item'): # Використовуємо forward reference 'Item'
        return self.value * other.value
    
    # 3. Змінено вивід __str__
    def __str__(self):
        return f"Екземпляр Item. Поточне значення: {self.value}. Статус: {self.status}"
    
    # 4. Змінено логіку __len__
    def __len__(self):
        return self.value // 10 + 1 # Повертає десятки + 1
    
    # 5. Змінено логіку staticmethod
    @staticmethod
    def calculate_rating():
        rating = 5
        if random.random() < 0.5:
            rating += 2
        else:
            rating -= 1
        return rating
    
    # 6. Змінено назву property
    @property
    def show_secret_key(self):
        return self.__secret_key

# --- Код, який використовує клас ---

c = Item()

# Доступ до публічного атрибута (Змінено ім'я з public_attribute на status)
print(f"Публічний атрибут: {c.status}")          

# Модифікація захищеного атрибута (Змінено ім'я з _private_attribute на _internal_id)
c._internal_id = f"ID-Modified-{c.value}" 
print(f"Захищений атрибут після модифікації: {c._internal_id}")         

# Вивід всіх атрибутів об'єкта
print("\n--- Вивід dir(c) ---")
print(dir(c))
print("--------------------")

# Доступ до приватного атрибута через property (Змінено назву property)
print(f"Приватний атрибут через property: {c.show_secret_key}") 

# Додаткова перевірка нових методів
d = Item()
print(f"Сума (використовується __add__, але логіка змінена на множення): {c + d}")
print(f"Довжина (використовується __len__): {len(c)}")
print(f"Статичний метод (викликає calculate_rating): {Item.calculate_rating()}")
```

Створив міні гру, де нам випадають ітеми або карти, якими можуть бути зброя, і ми робимо ходи цими картами, де карти атакують одна одну з attack_power та зменшують одна одній durability

```python
import random

class CardGame:
    CARD_NAMES = ["Dragon", "Knight", "Wizard", "Goblin", "Elf"]
    def __init__(self):
        self.name = random.choice(self.CARD_NAMES)
        self.attack_power = random.randint(5, 15)
        self.health = random.randint(20, 30)
    
    def attack(self, other_card): 
        other_card.health -= self.attack_power
        return f"{self.name} атакує {other_card.name}. {other_card.name} залишилось {other_card.health} здоров'я!"
    
    def defend(self):
        self.health += 2
        return f"{self.name} захищається, здоров'я відновилось до {self.health}!"
    
    def make_move(self, other_card):
        if random.choice([True, False]):
            return self.attack(other_card)
        else:
            return self.defend()
    
player_1 = CardGame()
print(f"Наша карта: {player_1.name} з атакою {player_1.attack_power} та здоровям {player_1.health}")

player_2 = CardGame()
print(f"Ворожа карта: {player_2.name} з атакою {player_2.attack_power} та здоровям {player_2.health}")

def shuffle_moves(player_1, player_2):
    if random.choice([True, False]):
        print(player_1.make_move(player_2))
        print(player_2.make_move(player_1))
    else:
        print(player_2.make_move(player_1))
        print(player_1.make_move(player_2))

for i in range(1, 10):
    print(f"Раунд {i}:")
    shuffle_moves(player_1, player_2)
    if player_1.health <= 0:
        print(f"{player_1.name} програв!")
        break
    elif player_2.health <= 0:
        print(f"{player_2.name} програв!")
        break
```

Модифікуємо нашу гру

```python
player_1 = [CardGame(), CardGame(), CardGame()]
for card in player_1:
    print(f"Наша карта: {card.name} з атакою {card.attack_power} та здоровям {card.health}")

player_2 = [CardGame(), CardGame(), CardGame()]
for card in player_2:
    print(f"Ворожа карта: {card.name} з атакою {card.attack_power} та здоровям {card.health}")

random.shuffle(player_1)
random.shuffle(player_2)

for i in range(1, 10):
    print(f"Раунд {i}:")
    card1 = random.choice(player_1)
    card2 = random.choice(player_2)
    print(card1.make_move(card2))
    print(card2.make_move(card1))

    if card1.health <= 0:
        print(f"{card1.name} програв!")
        player_1.remove(card1)
        print(f"Залишилось карт у гравця 1: {len(player_1)}")
    if card2.health <= 0:
        print(f"{card2.name} програв!")
        player_2.remove(card2)
        print(f"Залишилось карт у гравця 2: {len(player_2)}")
    
    if len(player_1) == 0:
        print("Гравець 1 програв! Всі наші карти програли!")
        break
    elif len(player_2) == 0:
        print("Гравець 2 програв! Всі ворожі карти програли!")
        break
```
