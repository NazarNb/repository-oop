## 1. Інкапсуляція
Інкапсуляція означає приховування деталей реалізації об'єкта, забезпечуючи доступ до них лише через спеціальні методи. В Python інкапсуляція реалізується за допомогою:

Публічних атрибутів і методів (public): 
- доступні звідусіль. Приватних атрибутів і методів (private): 
- позначаються символом _ або __ (наприклад, _attr, __attr).


```python
import random

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner  
        self.__balance = balance  

    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposited: {amount}")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn: {amount}")
        else:
            print(f"Attempted to withdraw {amount} - Insufficient funds")

    def get_balance(self):
        return self.__balance

account = BankAccount("Nazar", 1000)
for i in range(10):
    amount = random.randint(1, 500)  
    
    if random.choice([True, False]):
        account.deposit(amount)
    else:
        account.withdraw(amount)

print("\nFinal balance:", account.get_balance())
```

## 2.Наслідування
Створив ще один метод у класі Vehicle та викликав його з обєкта класу Car

```python
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def display_info(self):
        return f"{self.brand} {self.model}"
    
    def start_engine(self):
        return f"{self.brand} {self.model} engine started!"

class Car(Vehicle):
    def __init__(self, brand, model, seats):
        super().__init__(brand, model)  
        self.seats = seats

    def display_info(self):
        return f"{super().display_info()}, Seats: {self.seats}"

car = Car("Porshe", "911", 5)

print(car.display_info())
print(car.start_engine())  
```

## 3. Поліморфiзм
Створив клас Fish у якого немає методу speak, як результат виводить None

```python
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

class Fish(Animal):
    pass   # немає власного методу speak()


animals = [Dog(), Cat(), Fish()]

for animal in animals:
    print(animal.speak())
```

## 4. Абстракція
Абстракція приховує складність та залишає лише необхідні деталі. В Python це реалізується через абстрактні класи (використовуючи модуль abc).

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

circle = Circle(5)
print(circle.area())
```

5. Використання парадигм для створення простої гри

-  додано третій тип зброї Bow із власним параметром дальності та методом перезарядки 
- реалізовано заточку для сокири, запроваджено випадковий вибір зброї для гравця і бота також  створено покрокову інтерактивну систему бою з вибором дій користувачем,
- додано перевірку стану здоров’я для завершення гри та вдосконалено бойову механіку із застосуванням принципів абстракції, наслідування, інкапсуляції та поліморфізму.

```python
from abc import ABC, abstractmethod
from random import randint, choice

class Item(ABC):
    def __init__(self, name: str, health=500):
        self.name = name
        self.health = health
    
    @abstractmethod
    def attack(self, another_item):
        pass

    @abstractmethod
    def upgrade(self):
        pass

class Sword(Item):
    def __init__(self, name, attack_power: int):
        super().__init__(name=name)
        self.__attack_power = attack_power
        self._sharp = 0
    
    def attack(self, another_item: Item):
        current_attack = self.__attack_power + self._sharp + randint(0, 10)
        another_item.health -= current_attack
        return f"🗡️ Ви завдаєте удару мечем {self.name}! Шкода: {current_attack}. У {another_item.name} залишилось HP: {max(0, another_item.health)}"
    
    def upgrade(self):
        self._sharp += 5
        return f"✨ Ви гострите меч {self.name}. Бонус до атаки збільшено!"

class Axe(Item):
    def __init__(self, name, attack_power: int):
        super().__init__(name=name)
        self.__attack_power = attack_power
    
    def attack(self, another_item: Item):
        current_attack = self.__attack_power + randint(0, 25)
        another_item.health -= current_attack
        return f"🪓 Ви розмахуєте сокирою {self.name}! Шкода: {current_attack}. У {another_item.name} залишилось HP: {max(0, another_item.health)}"

    def upgrade(self):
        # Сокира важка, тому посилення дає випадковий великий бонус до наступного удару через тимчасову лють
        bonus = randint(5, 15)
        self.health += 10 # Сокира трохи лікує за рахунок бойового духу
        return f"😡 Ви входите в стан люті з {self.name}! HP відновлено на 10."

class Bow(Item): # Новий тип зброї
    def __init__(self, name, attack_power: int):
        super().__init__(name=name)
        self.__attack_power = attack_power
        self.range_power = 5
    
    def attack(self, another_item: Item):
        current_attack = self.__attack_power + randint(5, 15) + self.range_power
        another_item.health -= current_attack
        return f"🏹 Ви стріляєте з лука {self.name}! Шкода: {current_attack}. У {another_item.name} залишилось HP: {max(0, another_item.health)}"

    def reload(self): # Метод перезарядки/покращення дальності
        self.range_power += 3
        return f"⚙️ Ви калібруєте лук {self.name}. Дальність стрільби зросла!"
    
    def upgrade(self): # Реалізація абстрактного методу для інтерфейсу
        return self.reload()

# Функція для випадкового створення ворога
def create_random_weapon(name, power):
    weapon_type = choice([Sword, Axe, Bow])
    return weapon_type(name, power)

# Початок гри
print("🎮 Ласкаво просимо до бойової арени!")
player_name = input("Введіть ім'я свого героя: ")
player = create_random_weapon(player_name, 80)
enemy = create_random_weapon("Темний Лицар", 75)

print(f"\nВаша зброя: {type(player).__name__} '{player.name}'")
print(f"Ваш ворог: {type(enemy).__name__} '{enemy.name}'\n")

round_num = 1
while player.health > 0 and enemy.health > 0:
    print(f"--- РАУНД {round_num} ---")
    print(f"Ваше HP: {player.health} | HP Ворога: {enemy.health}")
    
    # Хід гравця
    print("Виберіть дію:")
    print("1. Атакувати")
    print("2. Посилити зброю")
    choice_action = input("Ваш вибір (1/2): ")

    if choice_action == "1":
        print(player.attack(enemy))
    else:
        print(player.upgrade())

    if enemy.health <= 0:
        print(f"\n🏆 ПЕРЕМОГА! {player.name} здолав ворога!")
        break

    # Хід ворога (автоматичний)
    enemy_action = randint(1, 2)
    if enemy_action == 1:
        print(enemy.attack(player))
    else:
        print(f"🌀 Ворог вирішив посилити свою зброю!")
        enemy.upgrade()

    if player.health <= 0:
        print(f"\n💀 ГРА ПРОВЕРШЕНА. {enemy.name} виявився сильнішим...")
        break

    round_num += 1
    print()
```
