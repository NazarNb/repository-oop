   - Приймає в конструкторі список чисел( наприклад,[1, -5, 2, 0, 6, 8, 3])
   - Має метод process(), який відсортовує список і залишає лише додатні значеняя
  - Має property result, яка повертає відсортований список додатних чисел

## User Promt

Твоє завдання наступне

- у кодовій комірці створити клас відповідно до завдання
- код повинен мати метод process(), який повинен відсортовувати і залишати лише додатні значення
- створити метод str  який буде виводити Ім'я та Групу студента, що виконує завдання( Ім'я -Назар, Група КН-31)
- Також код повинен містити property result , яка повертає відсортований список доданих чисел
- у циклі створити мінімум 2-3 об'єкти класу

```python
class PositiveNumbersSorter:
    def __init__(self, numbers, student_name, student_group):
        self.numbers = numbers
        self.student_name = student_name
        self.student_group = student_group
        self._result = []

    def process(self):
        self._result = sorted([n for n in self.numbers if n > 0])

    @property
    def result(self):
        return self._result

    def __str__(self):
        return f"Студент: {self.student_name}, Група: {self.student_group}"


data_sets = [
    [1, -5, 2, 0, 6, 8, 3],
    [-10, 4, 7, 1, 0, -3],
    [9, -2, -1, 5, 3]
]

student_name = "Назар Бобиляк"
student_group = "КН-31"

for data in data_sets:
    sorter = PositiveNumbersSorter(data, student_name, student_group)
    sorter.process()
    print(sorter)
    print("Вхідні дані:", data)
    print("Результат:", sorter.result)
    print("-" * 30)
```
